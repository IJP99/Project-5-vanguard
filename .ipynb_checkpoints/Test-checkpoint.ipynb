```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_demo = pd.read_csv("df_final_demo.txt")
df_f_exp = pd.read_csv("df_final_experiment_clients.txt")
df_web1 = pd.read_csv("df_final_web_data_pt_1.txt")
df_web2 = pd.read_csv("df_final_web_data_pt_2.txt")
```

```python
df_web = pd.concat([df_web1, df_web2], axis=0)
df_dupli = df_web.duplicated()
df_web = df_web.drop_duplicates()
```

```python
df_t = pd.merge(df_demo, df_f_exp, on="client_id")
```

```python
df = pd.merge(df_t, df_web, on="client_id")
```

```python
def column_name(data):
    data.columns = data.columns.str.replace(" ", "_")
    data.columns = data.columns.str.lower()
        
    return data
```
