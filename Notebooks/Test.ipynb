```python
df['date_time']= pd.to_datetime(df['date_time'])
def replace_confirm(df):
    return df.applymap(lambda x: "z_confirm" if x == "confirm" else x)
df = replace_confirm(df)
```

```python
df_test = df[df["variation"] == "Test"]
df_control = df[df["variation"] == "Control"]
```

```python
#completion rate comparison

print(df_test["process_step"].value_counts(normalize=True))

print(df_control["process_step"].value_counts(normalize=True))

#We can conclude that the test proccess population reach the confirm step in a 14.49% of the times while with the old version it was a 12.33% of the times
```

```python
step_order_control = (df_control['process_step'].shift(-1) < df_control['process_step']) | (df_control['process_step'].shift(-2) < df_control['process_step']) | (df_control['process_step'].shift(-3) < df_control['process_step']) | (df_control['process_step'].shift(-4) < df_control['process_step'])
df_control_errors = df_control[step_order_control]
number_control_erros = df_control_errors["client_id"].nunique()
number_id_control = df_control["client_id"].nunique()
prop_control_errors = number_control_erros/df_control["client_id"].nunique()
print(f"The number different client_id that we have in the test group is {number_id_control}")
print(f"The number of errors in this same group is {number_control_erros}")
print(f"the proportion is {prop_control_errors}")
```

```python
step_order_test = (df_test['process_step'].shift(-1) < df_test['process_step']) | (df_test['process_step'].shift(-2) < df_test['process_step']) | (df_test['process_step'].shift(-3) < df_test['process_step']) | (df_test['process_step'].shift(-4) < df_test['process_step'])
df_test_errors = df_test[step_order_test]
number_test_erros = df_test_errors["client_id"].nunique()
number_id_test = df_test["client_id"].nunique()
prop_test_errors = number_test_erros/df_test["client_id"].nunique()
print(f"The number different client_id that we have in the test group is {number_id_test}")
print(f"The number of errors in this same group is {number_test_erros}")
print(f"the proportion is {prop_test_errors}")
```

```python
#create the filter for test

df_test = df_test.sort_values(by=['client_id', 'date_time'])
step_order_filter = (df_test['process_step'].shift(-1) >= df_test['process_step']) & (df_test['process_step'].shift(-2) >= df_test['process_step']) & (df_test['process_step'].shift(-3) >= df_test['process_step']) & (df_test['process_step'].shift(-4) >= df_test['process_step'])
df_test = df_test[step_order_filter]
df_test = df_test.drop_duplicates(subset=['client_id', 'process_step'], keep='first')
df_pivot = df_test.pivot(index='client_id', columns='process_step', values='date_time')
df_pivot['time_start_to_step1'] = (df_pivot['step_1'] - df_pivot['start']).dt.total_seconds()
df_pivot['time_step1_to_step2'] = (df_pivot['step_2'] - df_pivot['step_1']).dt.total_seconds()
df_pivot['time_step2_to_step3'] = (df_pivot['step_3'] - df_pivot['step_2']).dt.total_seconds()
df_pivot['time_step3_to_confirm'] = (df_pivot['z_confirm'] - df_pivot['step_3']).dt.total_seconds()
df_pivot = df_pivot[
    (df_pivot['time_start_to_step1'] <= 3600) &
    (df_pivot['time_step1_to_step2'] <= 3600) &
    (df_pivot['time_step2_to_step3'] <= 3600) &
    (df_pivot['time_step3_to_confirm'] <= 3600) &
    (df_pivot["time_step3_to_confirm"] > 0)
]
```

```python
#create the filter for control

df_control = df_control.sort_values(by=['client_id', 'date_time'])
step_order_filter_control = (df_control['process_step'].shift(-1) >= df_control['process_step']) & (df_control['process_step'].shift(-2) >= df_control['process_step']) & (df_control['process_step'].shift(-3) >= df_control['process_step']) & (df_control['process_step'].shift(-4) >= df_control['process_step'])
df_control = df_control[step_order_filter_control]
df_control = df_control.drop_duplicates(subset=['client_id', 'process_step'], keep='first')
df_pivot_control = df_control.pivot(index='client_id', columns='process_step', values='date_time')
df_pivot_control['time_start_to_step1'] = (df_pivot_control['step_1'] - df_pivot_control['start']).dt.total_seconds()
df_pivot_control['time_step1_to_step2'] = (df_pivot_control['step_2'] - df_pivot_control['step_1']).dt.total_seconds()
df_pivot_control['time_step2_to_step3'] = (df_pivot_control['step_3'] - df_pivot_control['step_2']).dt.total_seconds()
df_pivot_control['time_step3_to_confirm'] = (df_pivot_control['z_confirm'] - df_pivot_control['step_3']).dt.total_seconds()
df_pivot_control = df_pivot_control[
    (df_pivot_control['time_start_to_step1'] <= 3600) &
    (df_pivot_control['time_step1_to_step2'] <= 3600) &
    (df_pivot_control['time_step2_to_step3'] <= 3600) &
    (df_pivot_control['time_step3_to_confirm'] <= 3600) &
    (df_pivot_control["time_step3_to_confirm"] > 0)
]
```

```python
#Since the new design (Test group) had a higher completion rate compared to the old design (Control group), you are required to confirm if this difference is statistically significant.
import scipy.stats as st

df_test['completed'] = df_test['process_step'].apply(lambda x: 1 if x == 'z_confirm' else 0)
df_control['completed'] = df_control['process_step'].apply(lambda x: 1 if x == 'z_confirm' else 0)
```

```python
#COMPLETION IN TEST
completions_control = df_control['completed']
total_control = df_control['client_id'].count()
completions_test = df_test['completed']
total_test = df_test['client_id'].count()
print(df_test[df_test["process_step"] == "z_confirm"]["process_step"].count())
print(df_control[df_control["process_step"] == "z_confirm"]["process_step"].count())
```

```python
counts = [completions_test, completions_control]
nobs = [total_test, total_control]
```

```python
st.ttest_ind(completions_control,completions_test, equal_var=False)

#We can say that the p-value is telling us to reject the null hipothesis of test and control being the same 
```

```python
control_completed_rate = ((completions_control.sum() / total_control)*100)
improve_5 = 1496.45
```

```python
#5% minimum improvement

st.ttest_1samp(completions_test, improve_5, alternative = "less")

#We can say that there is no possibility to have the 5% improvement
```
